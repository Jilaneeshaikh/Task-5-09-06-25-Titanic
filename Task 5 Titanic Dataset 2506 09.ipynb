```python
# Library Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Load Titanic Dataset
df = pd.read_csv("train.csv")
df.head()
print(df.tail())
```

```python
df.tail()
```

```python
df.shape
```

```python
# Basic Info
df.info()
df.describe()
df['Survived'].value_counts()
```

```python
# Pairplot
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

```

```python
# Count Plots
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Sex")
plt.show()

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# Age Distribution
df['Age'].hist(bins=30, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot: Age vs Survived
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survived")
plt.show()
```

```python
# Scatter Plot: Fare vs Age
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Fare vs Age (Colored by Survival)")
plt.show()
```

**Survival Rate:**

Overall survival rate is around 38%.

Women had a significantly higher survival rate than men.

First-class passengers were more likely to survive than second or third-class passengers.

**Numerical Correlations:**

Fare shows a moderate positive correlation with Survived  people who paid higher fares had better chances of survival.

Pclass has a negative correlation with Survived  higher-class passengers had better survival odds.

Age does not show a strong correlation, but younger passengers (especially children) had slightly better survival rates.

**Age Distribution:**

Most passengers were between 2040 years old.

Survivors were slightly younger on average.

**Sex and Class Influence:**

Females and 1st class passengers had the highest survival rates.

3rd class males had the lowest survival rate.

Fare Insights:**

There is a wide variation in fare amounts, with some outliers.

Higher fares generally correlate with better class and better survival chances.

Missing Values:

Age, Cabin, and Embarked columns contain missing values.

Cabin has many missing values and may not be directly useful without imputation or transformation.
